```python
# MONTE CARLO SIMULATION
import numpy as np

# Simulate 10,000 dice rolls
dice_rolls = np.random.randint(1, 7, size=10000)
print(dice_rolls[:20])  # showing first 20 values
```

```python
# Probability of each number
for i in range(1, 7):
    probability = np.sum(dice_rolls == i) / len(dice_rolls)
    print(f"Probability of {i}: {probability:.4f}")
```

```python
# Probability of getting an even number
even_prob = np.sum(dice_rolls % 2 == 0) / len(dice_rolls)
print("Probability of getting an even number:", even_prob)
```

```python
# Probability of getting at least one 6 in 10 rolls
experiments = np.random.randint(1, 7, size=(1000, 10))
atleast_one_six = np.sum(np.any(experiments == 6, axis=1)) / 1000
print("Probability of getting at least one 6 in 10 rolls:", atleast_one_six)
```

```python
import matplotlib.pyplot as plt

# Count how often each face appears
unique, counts = np.unique(dice_rolls, return_counts=True)

# Plot the bar chart
plt.figure()
plt.bar(unique, counts)
plt.xlabel("Dice Face (1â€“6)")
plt.ylabel("Frequency")
plt.title("Monte Carlo Simulation: Dice Roll Frequency (10,000 rolls)")
plt.show()
```

```python
# Count even vs odd outcomes
even_count = np.sum(dice_rolls % 2 == 0)
odd_count = len(dice_rolls) - even_count

# Data for visualization
labels = ["Even", "Odd"]
counts = [even_count, odd_count]

# Create Pie Chart
plt.figure()
plt.pie(counts, labels=labels, autopct="%0.2f%%")
plt.title("Probability Distribution: Even vs Odd (10,000 Dice Rolls)")
plt.show()
```

```python
# Run 1000 experiments, each experiment = 10 dice rolls
experiments = np.random.randint(1, 7, size=(1000, 10))

# Check if experiment contains at least one 6
has_six = np.any(experiments == 6, axis=1)

# Calculate probability
probability = np.sum(has_six) / 1000
print("Probability of getting at least one 6 in 10 rolls:", probability)

# Count how many had at least one 6 vs no 6
yes_count = np.sum(has_six)
no_count = 1000 - yes_count

# Visualization (Bar chart)
labels = ["At least one 6", "No 6"]
counts = [yes_count, no_count]

plt.figure()
plt.bar(labels, counts)
plt.xlabel("Outcome per experiment (10 rolls)")
plt.ylabel("Number of experiments")
plt.title("Monte Carlo Simulation: At least one 6 in 10 Rolls (1000 Experiments)")
plt.show()

```
